```python
import random as rnd
import math as m
import numpy as np
import statistics as stat
import scipy.stats as stat2
import time as tm
```

#Generación de aleatorios con el método de box muler

```python
N=100

z1=[]
z2=[]

t1=tm.time()
for i in range(N):
  u1= rnd.random()
  u2= rnd.random()
  R= m.sqrt(-2*m.log(u1))
  Th= 2*m.pi*u2
  x= R*m.cos(Th)
  y= R*m.sin(Th)
  z1.append(x)
  z2.append(y)
t2=tm.time()

t_muller=t2-t1
print("tiempo de ejecución "+ str(t_muller))
print(str(N) + " valores de la primera normal estándar")
print(z1)
print(str(N) + " valores de la segunda normal estándar")
print(z2)
```

#Prueba de Pearson

```python
categorias = round(1 + (m.log(len(z1)))/(m.log(2))) # Numero de categorias
maximo = max( z1 ) # Numero maximo de la lista generada
print( 'Categorias: {}'.format( categorias ) )
print( 'Maximo: {}'.format( maximo ) )

observados, bins = np.histogram( z1, categorias )
print( 'Observados: ', observados )
print( 'Bins: ', bins )

probs = stat2.norm.cdf(bins) # Funcion de distribucion (cdf)

print( 'Probs', probs )

P = []
for k in range( 1, categorias+1 ):
    cc = probs[k] - probs[k-1]
    P.append( cc )

print( 'P:', P )
print( 'Longitud P:', len(P) )

PROBS = np.array(P)
esperados = PROBS * len( z1 )
print( 'Esperados: ', esperados )
print( 'Sumatioria esperados: ', sum(esperados) )
print( 'Sumatoria observados: ', sum(observados) )

est_chi = 0
for k in range( categorias ):
    est_chi = est_chi + ((observados[k] - esperados[k])**2)/esperados[k]
print('est_chi: ', est_chi)

print(stat2.chi2.cdf(est_chi, categorias - 1))
```

Como el valor resultante de la prueba chi-cuadrado dio menor a 0.95 se acepta la hipotesis nula
