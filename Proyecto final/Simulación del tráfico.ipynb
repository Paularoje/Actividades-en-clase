```python
#@title Promedio de vehiculos presentes por carril ,      Funcion de Flujo (q) y Densidad (p)
def vehiculos_x_segundo():
  vc6=0
  vc7=0
  vc9=0
  vc10=0
  vf5=0
  vf6=0
  vf8=0
  vf9=0
####### Hallando el numero de vehiculos por segundo

  for i in range(0,len(data.data.loc[:,6])):
    if data.data.loc[i,6]==1:
      vc6+=1
  
  veh_c6.append(vc6)
  
  prom_vc6=stat.mean(veh_c6)
  
  for i in range(0,len(data.data.loc[:,7])):
    if data.data.loc[i,7]==1:
      vc7+=1
  
  veh_c7.append(vc7)
  
  prom_vc7=stat.mean(veh_c7)

  for i in range(0,len(data.data.loc[:,9])):
    if data.data.loc[i,9]==1:
      vc9+=1
  
  veh_c9.append(vc9)
  
  prom_vc9=stat.mean(veh_c9)

  for i in range(0,len(data.data.loc[:,10])):
    if data.data.loc[i,10]==1:
      vc10+=1
  
  veh_c10.append(vc10)
  
  prom_vc10=stat.mean(veh_c10)

  for i in range(0,len(data.data.loc[5])):
    if data.data.loc[5,i]==5:
      vf5+=1

  veh_f5.append(vf5)
  
  prom_vf5=stat.mean(veh_f5)

  for i in range(0,len(data.data.loc[6])):
    if data.data.loc[6,i]==5:
      vf6+=1

  veh_f6.append(vf6)
  
  prom_vf6=stat.mean(veh_f6)

  for i in range(0,len(data.data.loc[8])):
    if data.data.loc[8,i]==5:
      vf8+=1

  veh_f8.append(vf8)
  
  prom_vf8=stat.mean(veh_f8)

  for i in range(0,len(data.data.loc[9])):
    if data.data.loc[9,i]==5:
      vf9+=1
  
  veh_f9.append(vf9)
  
  prom_vf9=stat.mean(veh_f9)
  print('\n\nPromedio de Vehiculos por segundo:\n\nVehiculos en la columna 6: ',prom_vc6,'\nVehiculos en la columna 7: ',
      prom_vc7,'\nVehiculos en la columna 9: ',prom_vc9,'\nVehiculos en la columna 10: ',prom_vc10,'\nVehiculos en la fila 5: ',
      prom_vf5,'\nVehiculos en la fila 6: ',prom_vf6,'\nVehiculos en la fila 8: ',prom_vf8,'\nVehiculos en la fila 9: ',prom_vf9)

def flujo():
  q_c6=(cont_c6/tiempo)
  q_c7=(cont_c7/tiempo)
  q_c9=(cont_c9/tiempo)
  q_c10=(cont_c10/tiempo)
  q_f5=(cont_f5/tiempo)
  q_f6=(cont_f6/tiempo)
  q_f8=(cont_f8/tiempo)
  q_f9=(cont_f9/tiempo)
  q_f21=(cont_f21/tiempo)
  q_f32=(cont_f32/tiempo)
  q_33=(cont_f33/tiempo)
  print('\n\nFlujo de vehiculos en',tiempo,'segundos:\n\nFlujo columna 6: ',q_c6,'\nFlujo columna 7: ',q_c7,'\nFlujo columna 9: ',q_c9,'\nFlujo columna 10: ',q_c10,
      '\nFlujo fila 5: ',q_f5,'\nFlujo fila 6: ',q_f6,'\nFlujo fila 8: ',q_f8,'\nFlujo fila 9: ',q_f9)

def densidad():
  p_c6=cont_c6/len(data.data.loc[:,6])
  p_c7=cont_c7/len(data.data.loc[:,7])
  p_c9=cont_c9/len(data.data.loc[:,9])
  p_c10=cont_c10/len(data.data.loc[:,10])
  p_f5=cont_f5/len(data.data.loc[5])
  p_f6=cont_f6/len(data.data.loc[6])
  p_f8=cont_f8/len(data.data.loc[8])
  p_f9=cont_f9/len(data.data.loc[9])
  
  print('\n\nDensidad de vehiculos en',tiempo,'segundos:\n\nDensidad columna 6: ',p_c6,'\nDensidad columna 7: ',p_c7,'\nDensidad columna 9: ',p_c9,
        '\nDensidad columna 10: ',p_c10,'\nDensidad fila 5: ',p_f5,'\nDensidad fila 6: ',p_f6,'\nDensidad fila 8: ',p_f8,'\nDensidad fila 9: ',p_f9)
```
