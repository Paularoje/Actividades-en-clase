<h4><b>1. Definición de magnitudes de interés</b></h4>

<p>
  Se entiende por magnitud al tamaño o grandeza de un cuerpo. También puede indicar la importancia de una circunstancia o problema en particular. Por ejemplo, “Ese barco carguero es de una magnitud inimaginable”; “El huracán dejó destrozos de magnitud nunca antes vistos”.

En el área de las ciencias el término magnitud también es ampliamente utilizado, esto tiene que ver con que es una propiedad de los cuerpos, la cual puede ser medible y representada.

Teniendo en cuenta la definición de magnitud y el problema en particular, que en este caso es, simulación del tráfico en cierta ubicación de la ciudad de Bucaramanga, se establecieron las siguientes magnitudes de interés: 

<ul>
  <li>Tiempo de recorrido entre un sector A y un sector B</li>
  <li>Tiempo de aparicion de vehículos</li>
  <li>Cantidad de vehiculos presentes</li>
  <li>Tamaño de la ubicación estudiada</li>
  <li>Variación de la densidad del tráfico en un determinado periodo de tiempo</li> <li>Flujo del tráfico</li>
</ul>

<h4><b>4. Identificación de las reglas de comportamiento del sistema</b></h4>

<p>
   <br>
  Se toma cada iteracion como el transcurso de un segundo, en el que cada vehiculo se movera dependiendo de la direccion del carril (si es vertical o horizontal). Además, Se establecieron ciertas posiciones del dataframe como puntos de generacion de vehiculos y en el momento de un vehiculo pasar por una determinada esquina de un sector, generamos un valor aleatorio, en el cual, éste decide su direccion (si tomar a su derecha o seguir recto), dado el caso cambiara de direccion, se notaría visualmente, ya que, se tiene un color acorde a la orientacion que lleva el vehiculo.
</p>
