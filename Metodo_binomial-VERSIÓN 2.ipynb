```python
import numpy as np
import random
import math as m
import statistics as stat
import scipy.stats as stat2
```

# Generación de variable aleatoria binomial

```python
U = np.random.random(5) # Numero entero aleatorio (0 - 10)
n = len(U) # n = longitud de U
p = 0.4 # Probabilidad de exitos 

Xb = [] # Lista que guarda numeros aleatorios Binomial

for u in U:
  c = p/( 1-p )
  i = 0
  pr = ( 1-p )**n
  F = pr
  band = 0
  while ( band == 0 ):
    if u < F:
      Xb.append(i)
      band = 1
    else:
      pr = ( c*(n - i)/( i + 1 ) )*pr
      F = F + pr
      i = i + 1

print( Xb )
```

# Prueba de bondad de ajuste (Prueba de pearson)

```python
categorias = round(1 + (m.log(len(Xb)))/(m.log(2))) # Numero de categorias
maximo = max( Xb ) # Numero maximo de la lista generada
print( 'Categorias: {}'.format( maximo ) )

observados, bins = np.histogram( Xb, categorias )
print( 'Observados: ', observados )
print( 'Bins: ', bins )


# ¿Está bien esta formula?
probs = stat2.binom.cdf( bins, categorias, p ) # funcion de densidad


print( 'Probs', probs )

P = []
for k in range( 1, categorias+1 ):
    cc = probs[k] - probs[k-1]
    P.append( cc )

print( 'P:', P )
print( 'Longitud P:', len(P) )

PROBS = np.array(P)
esperados = PROBS * len( Xb )
print( 'Esperados: ', esperados )
print( 'Sumatioria esperados: ', sum(esperados) )
print( 'Sumatoria observados: ', sum(observados) )

est_chi = 0
for k in range( categorias ):
    est_chi = est_chi + ((observados[k] - esperados[k])**2)/esperados[k]
print('est_chi: ', est_chi)

result = stat2.chisquare(observados, esperados, categorias - 1, axis = None )
print('Result.statistic: ', result.statistic)
print(stat2.chi2.cdf(est_chi, categorias - 1))
print(stat2.chi2.cdf(result.statistic, categorias - 1))
```
